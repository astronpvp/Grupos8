1) Definir una función python llamada "simular_ruleta" que reproduzca n tiradas de una ruleta con numeros de 0 a 36.
- Lis


```
import random
import matplotlib.pyplot as plt
from collections import Counter

def simular_ruleta(n):
    """
    Simulate n spins of a roulette wheel (numbers 0-36) and plot a histogram.
    Also displays the numbers that appeared most and least frequently.

    Parameters:
    n (int): Number of spins to simulate

    Returns:
    list: List of spin results
    dict: Frequency count of each number
    """
    # European roulette has numbers 0-36
    spins = [random.randint(0, 36) for _ in range(n)]
    freq = Counter(spins)

    # Create histogram
    plt.figure(figsize=(12, 6))
    plt.hist(spins, bins=37, range=(-0.5, 36.5), density=True, edgecolor='black', alpha=0.7)
    plt.title(f'Roulette Spin Distribution (n={n:,} spins)')
    plt.xlabel('Number')
    plt.ylabel('Relative Frequency')
    plt.xticks(range(0, 37))
    plt.grid(axis='y', alpha=0.5)

    # Calculate and plot expected uniform distribution
    expected = 1/37
    plt.axhline(y=expected, color='red', linestyle='--', label='Expected Frequency (1/37)')
    plt.legend()

    # Find most and least frequent numbers
    most_common = freq.most_common(1)[0]
    least_common = freq.most_common()[-1]

    # Add text box with frequency information
    info_text = (f"Most frequent: Number {most_common[0]} (appeared {most_common[1]} times, {most_common[1]/n:.2%})\n"
                 f"Least frequent: Number {least_common[0]} (appeared {least_common[1]} times, {least_common[1]/n:.2%})")

    plt.gcf().text(0.15, 0.85, info_text, bbox=dict(facecolor='white', alpha=0.8),
                fontsize=10, verticalalignment='top')

    plt.show()

    # Print frequency information to console
    print(f"\nResults for {n:,} spins:")
    print(f"Most frequent number: {most_common[0]} (appeared {most_common[1]} times, {most_common[1]/n:.2%})")
    print(f"Least frequent number: {least_common[0]} (appeared {least_common[1]} times, {least_common[1]/n:.2%})")

    return spins, freq

# Run simulations with different sample sizes
sample_sizes = [100, 500, 1000, 5000, 10000]
results = {}

for size in sample_sizes:
    spins, freq = simular_ruleta(size)
    results[size] = {'spins': spins, 'frequency': freq}
    print("-" * 60)
```

```
def generar_poisson(n, l):
    return np.random.poisson(lam=l, size=n)
```

n: cantidad de valores que queremos generar

```
def mostrar_histograma(datos, titulo="Histograma"):
    plt.hist(datos, bins=range(min(datos), max(datos) + 2), edgecolor='black', align='left')
    plt.title(titulo)
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()
```

5) Definir una función de devuelva n valores de una distribución exponencial de scale 1/lambda.
Definir una función que imprima el histograma. Probar ambas funciones para varios tamaños y lambda=3.
Imprimir los valores máximos y mínimos obtenidos

Generar valores de una distribución exponencial

```
def generar_exponencial(n, lamb):
    scale = 1 / lamb
    return np.random.exponential(scale=scale, size=n)
```

¿Qué hace?

Genera n valores usando la distribución exponencial con media 1/λ.

Se usa scale = 1 / λ tal como pide el enunciado (en este caso, λ = 3 → scale = 1/3 ≈ 0.333).

 Paso 2: Mostrar el histograma y valores extremos

```
def mostrar_histograma_exponencial(datos, lamb, n):
    plt.hist(datos, bins=15, edgecolor='black')
    plt.title(f"Histograma Exponencial (n={n}, λ={lamb})")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

    print(f"Valor mínimo: {min(datos):.4f}")
    print(f"Valor máximo: {max(datos):.4f}")
```

 ¿Qué hace?

Muestra un histograma de los datos generados para visualizar su distribución.

Usa 15 intervalos (bins=15) para agrupar los valores.

Imprime también el valor mínimo y máximo observado, lo cual da una idea de la dispersión.

```
# 5) Distribución Exponencial
import numpy as np
import matplotlib.pyplot as plt

def generar_exponencial(n, lamb):
    """
    Genera n valores a partir de una distribución exponencial con media 1/lamb.
    """
    scale = 1 / lamb
    return np.random.exponential(scale=scale, size=n)

def mostrar_histograma_exponencial(datos, lamb, n):
    """
    Muestra un histograma para una muestra de datos exponenciales.
    """
    plt.hist(datos, bins=15, edgecolor='black')
    plt.title(f"Histograma Exponencial (n={n}, λ={lamb})")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

    print(f"Valor mínimo: {min(datos):.4f}")
    print(f"Valor máximo: {max(datos):.4f}")

# Ejemplo con n = 100 y lambda = 3
n = 100
lamb = 3
valores_exp = generar_exponencial(n, lamb)
mostrar_histograma_exponencial(valores_exp, lamb, n)
```

 ¿Qué se espera?

Que la mayoría de los valores estén cerca de 0, disminuyendo en frecuencia conforme aumentan.

Es característico de la distribución exponencial: mucha probabilidad de tiempos cortos y menos probabilidad de tiempos largos.

Prueba 1: n = 10 (muestra muy pequeña)

```
valores_exp = generar_exponencial(10, 3)
mostrar_histograma_exponencial(valores_exp, 3, 10)
```

Conclusión: Con una muestra tan pequeña, los resultados son muy variables. El histograma es irregular y no refleja bien la forma de la distribución exponencial.



Prueba 2: n = 500 (muestra media)

```
valores_exp = generar_exponencial(500, 3)
mostrar_histograma_exponencial(valores_exp, 3, 500)
```

 Conclusión: La forma exponencial comienza a definirse mejor. La mayoría de los valores siguen concentrándose cerca de cero, con algunos valores más altos hacia la derecha (cola larga).



 Prueba 3: n = 5000 (muestra grande)

```
valores_exp = generar_exponencial(5000, 3)
mostrar_histograma_exponencial(valores_exp, 3, 5000)
```

Conclusión: El histograma se ajusta muy bien a la forma esperada de una distribución exponencial. Se observa claramente una alta frecuencia de valores pequeños y una cola descendente hacia valores mayores.

Conclusión del ejercicio 5:
La muestra generada con distribución exponencial muestra una clara concentración de valores cerca de 0, con una larga cola hacia la derecha. Esto es típico cuando λ = 3. Al imprimir los valores mínimo y máximo, se observa la gran variabilidad posible, confirmando que esta distribución es útil para modelar tiempos entre eventos aleatorios.

6) Generar una muestra de tamaño 100 de 4 números enteros 0 a 3. Definir una función que dado el array y uno de los números, devuelva el ratio de ocurrencia.

Paso 1: Función para calcular el ratio de ocurrencia

```
def ratio_ocurrencia(array, valor):
    return np.count_nonzero(array == valor) / len(array)
```

¿Qué hace?

Cuenta cuántas veces aparece un valor específico en un array (np.count_nonzero(array == valor)).

Divide esa cantidad por el tamaño total del array para obtener el porcentaje de ocurrencia (ratio).

 Paso 2: Generar muestra aleatoria


```
muestra = np.random.randint(0, 4, size=100)
```

¿Qué hace?

Genera una muestra de tamaño 100 con números enteros entre 0 y 3 (inclusive).

Los valores se distribuyen de forma uniforme por defecto.



Paso 3: Calcular el ratio de ocurrencia de un valor

```
valor = 2
r = ratio_ocurrencia(muestra, valor)
print(f"El número {valor} aparece con un ratio de: {r:.2%}")
```

¿Qué hace?

Calcula qué porcentaje de veces aparece el valor 2 en la muestra.

El resultado se imprime como porcentaje con dos decimales.

Paso 4: Mostrar el histograma de la muestra


```
plt.hist(muestra, bins=4, edgecolor='black')
plt.title("Histograma de muestra (valores 0 a 3)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

¿Qué hace?

Dibuja un histograma que muestra cuántas veces aparece cada valor (0, 1, 2 y 3).

Sirve para visualizar si los valores están distribuidos de forma equilibrada o no.



```
# 6) Generar muestra y calcular ratio de ocurrencia

def ratio_ocurrencia(array, valor):
    """
    Calcula el ratio de ocurrencia de un valor dentro de un array.
    """
    return np.count_nonzero(array == valor) / len(array)

# Generar muestra de tamaño 100 con valores entre 0 y 3
muestra = np.random.randint(0, 4, size=100)
print("Muestra generada:")
print(muestra)

# Calcular el ratio de ocurrencia para el valor 2
valor = 2
r = ratio_ocurrencia(muestra, valor)
print(f"El número {valor} aparece con un ratio de: {r:.2%}")

# Mostrar histograma de la muestra
plt.hist(muestra, bins=4, edgecolor='black')
plt.title("Histograma de muestra (valores 0 a 3)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

7) Generar una muestra de opciones 'a','b','c','d' con diferentes pesos. Probar distintos casos y graficar histograma


Definimos una función para generar opciones con distintas probabilidades

```
def generar_muestra_opciones(opciones, pesos, n):
    return np.random.choice(opciones, size=n, p=pesos)
```

Este codigo utiliza np.random.choice para seleccionar n elementos aleatorios de la lista opciones.

Elige cada opción según los pesos dados en pesos, que indican la probabilidad de que salga cada una.

Es importante que los pesos sumen 1, si no, da error.

* PASO 2 CREAR UNA FUNCIÓN PARA MOSTRAR EL HISTOGRAMA

```
def mostrar_histograma_opciones(datos, titulo="Histograma de opciones"):
    plt.hist(datos, bins=np.arange(len(set(datos)) + 1) - 0.5, edgecolor='black', align='mid', rwidth=0.8)
    plt.xticks(range(len(set(datos))), sorted(set(datos)))
    plt.title(titulo)
    plt.xlabel("Opción")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()
```

-Esta función dibuja un histograma de los valores categóricos ('a','b',...)

-usa bins con desplazamieno para que cada letra quede centrada



Probamos con un ejemplo real:

```
opciones = ['a', 'b', 'c', 'd']
pesos = [0.1, 0.2, 0.3, 0.4]  # Probabilidad de cada opción
n = 1000

muestra = generar_muestra_opciones(opciones, pesos, n)
mostrar_histograma_opciones(muestra, "Muestra con pesos [0.1, 0.2, 0.3, 0.4]")
```

 ¿Qué se espera?

Que la opción 'd' aparezca más veces que las demás porque tiene mayor peso (0.4).

Que la opción 'a' aparezca pocas veces (peso de 0.1).

El histograma te lo muestra visualmente.

EJEMPLO CON PESOS IGUALES

```
opciones = ['a', 'b', 'c', 'd']
pesos = [0.25, 0.25, 0.25, 0.25]
n = 1000

muestra = generar_muestra_opciones(opciones, pesos, n)
mostrar_histograma_opciones(muestra, "Muestra con pesos iguales [0.25, 0.25, 0.25, 0.25]")
```

Aquí esperamos que todas las letras aparezcan más o menos la misma cantidad de veces, ya que tienen la misma probabilidad.

* EJEMPLO CON PESOS EXTREMOS:

```
opciones = ['a', 'b', 'c', 'd']
pesos = [0.7, 0.1, 0.1, 0.1]
n = 1000

muestra = generar_muestra_opciones(opciones, pesos, n)
mostrar_histograma_opciones(muestra, "Muestra con peso dominante [0.7, 0.1, 0.1, 0.1]")
```

Aquí esperamos que 'a' aparezca con mucha más frecuencia que el resto.



Conclusion ejercicio 7:

Al generar muestras con diferentes pesos, observamos cómo la probabilidad asignada a cada opción se refleja en el histograma:

Con pesos iguales, todas las opciones aparecen con frecuencia similar.

Con pesos desiguales, las opciones con mayor peso aparecen más frecuentemente.
Esto confirma que el parámetro p en np.random.choice controla efectivamente la distribución de la muestra.
